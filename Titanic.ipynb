```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Acquire Data
titanic_data = pd.read_csv('data/titanic_dataset.csv')

#Normalize Data
```

```python
#Draw heatmap
import seaborn as sns

dims = (11.7,8.27)
fig,ax = plt.subplots(figsize=dims)
CorMap = sns.heatmap(titanic_data.corr(), annot = True,fmt= '.1f', ax=ax)
plt.title('Titanic Data Correlation')
```

```python
#Implementing a stradified struffle split to split the data 80/20
#Change this to random to meet assignment requirements or state that this version is better
 
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2)
for train_indices, test_indices in split.split(titanic_data, titanic_data[["Survived", "Pclass"]]):
                                                                              strat_train_set = titanic_data.loc[train_indices]
                                                                              strat_test_set = titanic_data.loc[test_indices]
```

```python
plt.subplot(1,2,1)
strat_train_set['Survived'].hist()
strat_train_set['Pclass'].hist()

plt.show
```

```python
plt.subplot(1,2,1)
strat_test_set['Survived'].hist()
strat_test_set['Pclass'].hist()

plt.show
```
